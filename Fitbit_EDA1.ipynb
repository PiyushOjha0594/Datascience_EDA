```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importing Dataset
fitbit = pd.read_csv(r'D:\Dataset\FitBit\FitBit data.csv')
fitbit
```

```python
#checking the number of rows and columns in the dataset
fitbit.shape
```

```python
#Checking for the null value in the dataset
fitbit.isnull().sum() 
```

```python
#sample first 5 rows from dataset
fitbit.head()
```

```python
#copying the dataset to fitbit1 frame
fitbit1 =fitbit.copy()
```

```python
# adding the yearm month and date columns to the dataset
fitbit1["year"] = pd.DatetimeIndex(fitbit1["ActivityDate"]).year
fitbit1["month"] = pd.DatetimeIndex(fitbit1["ActivityDate"]).month
fitbit1["date"] = pd.DatetimeIndex(fitbit1["ActivityDate"]).day
```

```python
#sample first 5 rows from dataset after updating the year,month abd date column
fitbit1.head()
```

```python
plt.figure(figsize =(15,8))
ax =sns.boxplot(x="date", y ="Calories",data = fitbit1)
ax.set_title('Box plot of Calories date of the month')
```

```python
plt.figure(figsize = (15,8))
sc = sns.scatterplot(x ='Calories',y  = 'SedentaryMinutes',data = fitbit1)
sc.set_title('Scatterplot of calories and SedentaryMinutes')
```

```python
plt.figure(figsize = (15,8))
st = sns.scatterplot(x ='Calories',y  = 'LightlyActiveMinutes',data = fitbit1)
st.set_title('Scatterplot of calories and LightlyActiveMinutes')
```

```python
plt.figure(figsize = (15,8))
sr = sns.scatterplot(x ='Calories',y  = 'FairlyActiveMinutes',data = fitbit1)
sr.set_title('Scatterplot of calories and FairlyActiveMinutes')
```

```python
all_col = ['LightActiveDistance','VeryActiveMinutes','SedentaryMinutes','LightlyActiveMinutes','FairlyActiveMinutes','Calories']
df = fitbit1[all_col]
plt.figure(figsize=(15,8))
ca=sns.scatterplot(data = df )
ca.set_title('Scatterplot of calories and with minutes')
```

```python
#plotting with different activities with distance
all_distance = ['TotalDistance','TrackerDistance','LoggedActivitiesDistance','VeryActiveDistance','ModeratelyActiveDistance','LightActiveDistance']
Distance = fitbit1[all_distance]
dis=sns.catplot(data = Distance,height =(15))
ca.set_title('Scatterplot of calories and with Distance')
```
